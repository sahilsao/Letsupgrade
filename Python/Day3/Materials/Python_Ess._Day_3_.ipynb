# Control Statements / Keywords

```python
# Break - it is been used for breaking the existing for loop

# Pass - Placeholder 

# Continue - it is used for Skipping a particular statement
```

```python
for item in range(1,10):
    print(item)
    if item == 5:
        break;
```

```python
for item in range(1,10):
    if item == 5:
        continue;
    print(item)
```

```python
for item in range(1,10):
```

```python
for item in range(1,10):
    pass
```
